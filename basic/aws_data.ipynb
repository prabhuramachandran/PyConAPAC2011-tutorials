## Date and time in Python

```python
date0 = data['dateist'][100]
date0
```

```python
time0 = data['timeist'][0]
time0
```

Notice that the time has the wrong date.  Turns out that the month and day are also incorrectly parsed.  Fortunately, csv2rec supports day-first reading.

```python
data = csv2rec('aws_data_clean.csv', dayfirst=True)
date0 = data['dateist'][100]
time0 = data['timeist'][0]
```

```python
date0.year, date0.month, date0.day
```

```python
new_time = time0.replace(month=date0.month, day=date0.month, year=date0.year)
new_time
```

```python
delta = new_time - time0
delta
```

With this, we can create the correct time data.

## Generating the right time sequence

Loop over data and create a new list with the time data.

Note that to loop over a range of indices use `range(n)`.

```python
time = []
for i in range(len(data)):
    d = data['dateist'][i]
    t = data['timeist'][i]
    new_t = t.replace(year=d.year, month=d.month, day=d.day)
    time.append(new_t)
```

```python
time = asarray(time)
time[0]
```

```python
plot(time, data['rain_fall'])
xticks(rotation=45)
grid();
```

```python
plot(time, data['atmo_pressure'])
xticks(rotation=45);
```

```python
plot(time, data['air_temp'])
xticks(rotation=45);
```

# Using pandas for the above

Using pandas makes some of this easier as it is naturally suited for time-series data.

Pandas has a `date_range` function.

```python
import pandas as pd
dates = pd.date_range(start='2013-06-01 05:30:00+05:30', freq="H", periods=len(data))
```

```python
df = pd.DataFrame(data={'rain_fall':data['rain_fall'], 
                        'air_temp':data['air_temp'], 
                        'wind_speed':data['wind_speed'],
                        'atmo_pressure':data['atmo_pressure']}, index=dates)
```

## Interpolation of data

```python
from scipy.interpolate import interp1d
```

```python
idx = ravel(indices(time.shape))
sfit = interp1d(idx, data['rain_fall'], kind=2)
plot(idx, sfit(idx))
plot(idx, data['rain_fall'])
```

```python
sfit(1300)
```

```python
from numpy import polyfit, poly1d
coeffts = polyfit(idx, data['rain_fall'], 10)
pfit = poly1d(coeffts)
plot(idx, data['rain_fall'])
plot(idx, pfit(idx)) 
```
